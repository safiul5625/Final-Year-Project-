```python
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Calculate the Pearson correlation matrix for the combined DataFrame.
correlation_matrix = df.corr(method='pearson')

# 3. Extract the correlations of all features with the target column.
correlations = correlation_matrix['binary_sleep'].drop('binary_sleep')

# 4. Identify features where the absolute correlation with target
high_corr_features = correlations[abs(correlations) > 0.1].index.tolist()

# 5. Filter X_scaled_resampled to include only these highly correlated features
X_selected_features = X_scaled_resampled[high_corr_features]


# 6. Generate a heatmap of the correlation matrix for the X_selected_features and y_resampled
plt.figure(figsize=(10, 8))
sns.heatmap(
    pd.concat([X_selected_features, y_resampled.rename('binary_sleep')], axis=1).corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title('Correlation Heatmap of Selected Features with Hours of sleep')
plt.show()
```

```python
##Test-Train

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_selected_features, y_resampled, test_size=0.2, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn.model_selection import  cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def train_evaluate_model(classifier, X_train, y_train, X_test, y_test, model_name):
    print(f"\n--- {model_name} ---")

    # a. Fit the model to the training data
    classifier.fit(X_train, y_train)

    # b. Make predictions on both the training data and the test data
    y_train_pred = classifier.predict(X_train)
    y_test_pred = classifier.predict(X_test)

    # c. Calculate and print the training accuracy and test accuracy
    train_accuracy = accuracy_score(y_train, y_train_pred)
    Atrain.append(train_accuracy)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    Atest.append(test_accuracy)
    print(f"Training Accuracy: {train_accuracy:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")

    # d. Perform 5-fold cross-validation on X_train and y_train
    cv_scores = cross_val_score(classifier, X_train, y_train, cv=5, scoring='accuracy')
    Acv.append(np.mean(cv_scores))
    print(f"Mean 5-fold Cross-Validation Accuracy: {np.mean(cv_scores):.4f}")
    print(f"Cross-Validation Scores: {cv_scores}")

    # Store model name
    model_names.append(model_name)

    # Calculate ROC metrics and store them
    # Check if the classifier has predict_proba method (needed for ROC curve)
    if hasattr(classifier, "predict_proba"):
        y_test_proba = classifier.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_test_proba)
        roc_auc = roc_auc_score(y_test, y_test_proba)
        all_fpr.append(fpr)
        all_tpr.append(tpr)
        all_roc_auc.append(roc_auc)
        print(f"ROC AUC: {roc_auc:.4f}")
    else:
        # For classifiers like SVC with default settings, predict_proba might not be available
        # if not explicitly set (e.g., SVC(probability=True))
        all_fpr.append(None)
        all_tpr.append(None)
        all_roc_auc.append(None)
        print("ROC AUC: N/A (predict_proba not available or probability=False for SVC)")

    # e. Generate and display a confusion matrix for the test set predictions
    cm = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # f. Print a classification report for the test set predictions
    print(f"\nClassification Report for {model_name}:\n")
    print(classification_report(y_test, y_test_pred))
```

```python
##Base random forest

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
train_evaluate_model(
    rf,
    X_train,
    y_train,
    X_test,
    y_test,
    "Base RandomForest Classifier"
)
```

```python
import matplotlib.pyplot as plt

model_name_to_plot = "Base RandomForest Classifier" # Use the exact name used in train_evaluate_model

if model_name_to_plot in model_names:
    model_index = model_names.index(model_name_to_plot)

    # Retrieve the ROC data for the specific model
    fpr_rf = all_fpr[model_index]
    tpr_rf = all_tpr[model_index]
    roc_auc_rf = all_roc_auc[model_index]

    if fpr_rf is not None and tpr_rf is not None and roc_auc_rf is not None:
        # Create the plot
        plt.figure(figsize=(8, 6))
        plt.plot(fpr_rf, tpr_rf, color='blue', lw=2, label=f'{model_name_to_plot} (AUC = {roc_auc_rf:.4f})')
        plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--', label='Random Classifier')

        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate', fontsize=12)
        plt.ylabel('True Positive Rate', fontsize=12)
        plt.title(f'ROC Curve for {model_name_to_plot}', fontsize=14, fontweight='bold')
        plt.legend(loc="lower right", fontsize=10)
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()
    else:
        print(f"ROC data not available for '{model_name_to_plot}'.")
else:
    print(f"Model '{model_name_to_plot}' not found in the list of trained models.")
```

```python
# Calculate and print training and test error for Base RandomForest
# Training Error = 1 - Training Accuracy
# Test Error = 1 - Test Accuracy

# Use the exact name used in train_evaluate_model
model_name_for_error = 'Base RandomForest Classifier'

if model_name_for_error in model_names:
    model_index = model_names.index(model_name_for_error)

    base_rf_train_accuracy = Atrain[model_index]
    base_rf_test_accuracy = Atest[model_index]

    base_rf_train_error = 1 - base_rf_train_accuracy
    base_rf_test_error = 1 - base_rf_test_accuracy

    print(f"Base RandomForest Training Error (1 - Accuracy): {base_rf_train_error:.4f}")
    print(f"Base RandomForest Test Error (1 - Accuracy): {base_rf_test_error:.4f}")
else:
    print(f"Model '{model_name_for_error}' not found in the list of trained models.")
```

```python
##Tuned Random forest


from sklearn.model_selection import GridSearchCV

# Define the parameter grid for RandomForestClassifier
param_grid = {
    'n_estimators': [50, 100, 200], # Number of trees in the forest
    'max_depth': [None, 10, 20, 30], # Maximum depth of the tree
    'min_samples_split': [2, 5, 10], # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4] # Minimum number of samples required to be at a leaf node
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42), param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='accuracy')

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Get the best estimator
best_rf = grid_search.best_estimator_

print(f"Best parameters for RandomForest: {grid_search.best_params_}")

# Evaluate the best Random Forest model
train_evaluate_model(
    best_rf,
    X_train,
    y_train,
    X_test,
    y_test,
    "Tuned RandomForest Classifier"
)
```

```python
# coding: utf
"""Complete Model Evaluation - Train & Test ROC Together, Separate Loss & Accuracy Plots"""

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import roc_curve, auc, roc_auc_score, precision_score, recall_score, f1_score, log_loss
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Storage for metrics
Atrain = []
Atest = []
Acv = []
roc_auc_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
train_losses = []
test_losses = []

# Mapping dictionary
mapping = {
    "Age": {"20 – 24 years  (২০ – ২৪ বছর)": 0, "25 – 29 years  (২৫ – ২৯ বছর)": 1,
            "15 – 19 years  (১৫ – ১৯ বছর)": 2, "10 – 14 years  (১০ – ১৪ বছর)": 3},
    "Gender": {"Male — পুরুষ": 0, "Female — মহিলা": 1, "Prefer not to say — উত্তর দিতে ইচ্ছুক নই": 2},
    "Occupation": {"Student – ছাত্র": 0, "Working Professional –  কর্মজীবী": 1, "Retired –  অবসরপ্রাপ্ত": 2},
    "Do you smoke": {"No – না": 0, "Yes –  হ্যাঁ": 1, "Passive smoker – পরোক্ষ ধূমপায়ী": 2},
    "No of Ciggerates": {"I do not smoke – আমি ধূমপান করি না।": 0, "1–5 (১ – ৫)": 1, "6–10 (৬ – ১০)": 2,
                         "More than 15 – ১৫টির বেশি": 3, "11–15 (১১ – ১৫)": 4, "Irregular (sometimes 4/5 days gap)": 5, " ": 6},
    "Year of Smoking": {"I do not smoke – আমি ধূমপান করি না।": 0, "1–3 years – ১-৩ বছর": 1,
                        "4–6 years – ৪-৬ বছর": 2, "More than 7 years – ৭ বছরের বেশি": 3, "Less than 1 year – ১ বছরের কম": 4, " ": 5},
    "Time of smoking": {"I do not smoke – আমি ধূমপান করি না।": 0, "No fixed time – কোন নির্দিষ্ট সময় নেই": 1,
                        "Evening – সন্ধ্যা": 2, "Before bedtime – ঘুমানোর আগে": 3, "Afternoon – বিকেল": 4,
                        "Morning – সকাল": 5, "Morning evening and before bedtime": 6, " ": 7},
    "Trying to quite": {"No – না": 0, "Yes – হ্যাঁ": 1, "I do not smoke – আমি ধূমপান করি না।": 2},
    "Hours of sleep": {"6–8 (৬-৮)": 0, "4–6 (৪-৬)": 1, "Less than 4 – ৪ এর কম": 2, "More than 8 – ৮ এর বেশি": 3},
    "Hour of trying to sleep": {"30–60 minutes – ৩০-৬০ মিনিট": 0, "More than 1 hour – ১ ঘন্টার বেশি": 1,
                                "Less than 15 minutes – ১৫ মিনিটের কম": 2, "15–30 minutes – ১৫-৩০ মিনিট": 3},
    "wake up during sleep": {"Sometimes – মাঝে মাঝে": 0, "Never – কখনই না": 1, "Rarely – কদাচিৎ": 2,
                             "Every night – প্রতি রাতে": 3, "Often – প্রায়ই": 4},
    "Quality of sleep": {"Never – কখনই না": 0, "Rarely – কদাচিৎ": 1, "Sometimes – মাঝে মাঝে": 2,
                        "Often – প্রায়ই": 3, "Always – সর্বদা": 4},
    "Smoke before sleep": {"I do not smoke – আমি ধূমপান করি না।": 0, "I do not smoke before bed – আমি ঘুমানোর আগে ধূমপান করি না।": 1,
                          "Less than 30 minutes – ৩০ মিনিটের কম": 2, "30-60 minutes – ৩০-৬০ মিনিট": 3, "More than 60 minutes – ৬০ মিনিটের বেশি": 4},
    "caffeine after 6 PM": {"No – না": 0, "Yes – হ্যাঁ": 1},
    "exercise ": {"No – না": 0, "Yes – হ্যাঁ": 1},
    "anxiety": {"Rarely – কদাচিৎ": 0, "Sometimes – মাঝে মাঝে": 1, "Often – প্রায়ই": 2, "Always – সর্বদা": 3, "Never – কখনই না": 4},
    " sleep disorder": {"No – না": 0, "Yes – হ্যাঁ": 1, "Not sure – নিশ্চিত নই": 2}
}


def train_evaluate_model_complete(classifier, X_train, y_train, X_test, y_test, model_name):
    """
    Complete training and evaluation with:
    1. ROC Curve (Train & Test together)
    2. Separate Loss plot
    3. Separate Accuracy plot
    """
    print(f"\n{'='*90}")
    print(f"{'='*90}")
    print(f"  {model_name.upper()}")
    print(f"{'='*90}")
    print(f"{'='*90}\n")

    # Fit the model
    classifier.fit(X_train, y_train)

    # Predictions
    y_train_pred = classifier.predict(X_train)
    y_test_pred = classifier.predict(X_test)

    # Get probability predictions
    y_train_proba = None
    y_test_proba = None

    if hasattr(classifier, "predict_proba"):
        y_train_proba = classifier.predict_proba(X_train)
        y_test_proba = classifier.predict_proba(X_test)
    elif hasattr(classifier, "decision_function"):
        from scipy.special import expit
        y_train_scores = classifier.decision_function(X_train)
        y_test_scores = classifier.decision_function(X_test)
        y_train_proba = np.column_stack([1 - expit(y_train_scores), expit(y_train_scores)])
        y_test_proba = np.column_stack([1 - expit(y_test_scores), expit(y_test_scores)])

    # Calculate accuracies
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    Atrain.append(train_accuracy)
    Atest.append(test_accuracy)

    # Calculate loss
    train_loss = 0
    test_loss = 0
    if y_train_proba is not None:
        try:
            train_loss = log_loss(y_train, y_train_proba)
            test_loss = log_loss(y_test, y_test_proba)
        except:
            pass
    train_losses.append(train_loss)
    test_losses.append(test_loss)

    # Additional metrics
    precision = precision_score(y_test, y_test_pred, average='binary', zero_division=0)
    recall = recall_score(y_test, y_test_pred, average='binary', zero_division=0)
    f1 = f1_score(y_test, y_test_pred, average='binary', zero_division=0)

    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)

    # Print metrics
    print(f"📊 Training Accuracy:   {train_accuracy:.4f}")
    print(f"📊 Test Accuracy:       {test_accuracy:.4f}")
    print(f"📉 Training Loss:       {train_loss:.4f}")
    print(f"📉 Test Loss:           {test_loss:.4f}")
    print(f"📊 Precision:           {precision:.4f}")
    print(f"📊 Recall:              {recall:.4f}")
    print(f"📊 F1-Score:            {f1:.4f}")

    # Cross-validation
    cv_scores = cross_val_score(classifier, X_train, y_train, cv=5, scoring='accuracy')
    mean_cv_score = np.mean(cv_scores)
    std_cv_score = np.std(cv_scores)
    Acv.append(mean_cv_score)
    print(f"📊 Mean CV Accuracy:    {mean_cv_score:.4f} (±{std_cv_score:.4f})")

    # Calculate ROC curves
    train_roc_auc = 0
    test_roc_auc = 0
    train_fpr, train_tpr = [0, 1], [0, 1]
    test_fpr, test_tpr = [0, 1], [0, 1]

    if y_train_proba is not None and y_test_proba is not None:
        try:
            # Train ROC
            train_roc_auc = roc_auc_score(y_train, y_train_proba[:, 1])
            train_fpr, train_tpr, _ = roc_curve(y_train, y_train_proba[:, 1])

            # Test ROC
            test_roc_auc = roc_auc_score(y_test, y_test_proba[:, 1])
            test_fpr, test_tpr, _ = roc_curve(y_test, y_test_proba[:, 1])

            print(f"📊 Train ROC-AUC:       {train_roc_auc:.4f}")
            print(f"📊 Test ROC-AUC:        {test_roc_auc:.4f}")
        except Exception as e:
            print(f"📊 ROC-AUC:             N/A")

    roc_auc_scores.append(test_roc_auc)

    # ============= CREATE SEPARATE PLOTS FOR THIS MODEL =============

    # Figure 1: ROC Curve (Train & Test together)
    fig1 = plt.figure(figsize=(10, 8))
    if y_train_proba is not None and y_test_proba is not None:
        plt.plot(train_fpr, train_tpr, color='blue', lw=3,
                label=f'Train ROC (AUC = {train_roc_auc:.4f})')
        plt.plot(test_fpr, test_tpr, color='darkorange', lw=3,
                label=f'Test ROC (AUC = {test_roc_auc:.4f})')
        plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--',
                label='Random Classifier')
        plt.fill_between(test_fpr, test_tpr, alpha=0.2, color='darkorange')
    else:
        plt.text(0.5, 0.5, 'ROC Curve Not Available\n(Model does not support probabilities)',
                ha='center', va='center', fontsize=14,
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=14, fontweight='bold')
    plt.ylabel('True Positive Rate', fontsize=14, fontweight='bold')
    plt.title(f'{model_name} - ROC Curve (Train & Test)', fontsize=16, fontweight='bold', pad=20)
    plt.legend(loc="lower right", fontsize=12, frameon=True, shadow=True)
    plt.grid(alpha=0.4, linestyle='--')
    plt.tight_layout()
    plt.show()

    # Figure 2: Loss Comparison (Separate)
    fig2 = plt.figure(figsize=(10, 6))
    if train_loss > 0 and test_loss > 0:
        categories = ['Training Loss', 'Test Loss']
        values = [train_loss, test_loss]
        colors = ['#3498db', '#e74c3c']
        bars = plt.bar(categories, values, color=colors, edgecolor='black', linewidth=2, width=0.5)

        plt.ylabel('Log Loss', fontsize=14, fontweight='bold')
        plt.title(f'{model_name} - Loss Comparison', fontsize=16, fontweight='bold', pad=20)
        plt.grid(axis='y', alpha=0.4, linestyle='--')

        # Add value labels
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.4f}', ha='center', va='bottom',
                    fontsize=13, fontweight='bold')

        # Add info text
        plt.text(0.5, 0.95, f'Lower is Better | Difference: {abs(train_loss - test_loss):.4f}',
                ha='center', transform=plt.gca().transAxes, fontsize=11,
                bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))
    else:
        plt.text(0.5, 0.5, 'Loss Not Available', ha='center', va='center', fontsize=14,
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

    plt.tight_layout()
    plt.show()

    # Figure 3: Accuracy Comparison (Separate)
    fig3 = plt.figure(figsize=(10, 6))
    categories = ['Training\nAccuracy', 'Test\nAccuracy', 'CV\nAccuracy']
    values = [train_accuracy, test_accuracy, mean_cv_score]
    colors = ['#2ecc71', '#e74c3c', '#9b59b6']
    bars = plt.bar(categories, values, color=colors, edgecolor='black', linewidth=2, width=0.5)

    plt.ylabel('Accuracy', fontsize=14, fontweight='bold')
    plt.title(f'{model_name} - Accuracy Comparison', fontsize=16, fontweight='bold', pad=20)
    plt.ylim([0, 1.1])
    plt.grid(axis='y', alpha=0.4, linestyle='--')
    plt.axhline(y=1.0, color='red', linestyle='--', linewidth=1, alpha=0.5, label='Perfect Score')

    # Add value labels
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                f'{height:.4f}', ha='center', va='bottom',
                fontsize=13, fontweight='bold')

    # Add overfitting check
    overfit_gap = train_accuracy - test_accuracy
    overfit_status = "⚠️ OVERFITTING!" if overfit_gap > 0.1 else "✓ Good Fit"
    plt.text(0.5, 0.95, f'{overfit_status} | Train-Test Gap: {overfit_gap:.4f}',
            ha='center', transform=plt.gca().transAxes, fontsize=11,
            bbox=dict(boxstyle='round', facecolor='lightcoral' if overfit_gap > 0.1 else 'lightgreen', alpha=0.8))

    plt.legend(loc='upper right', fontsize=10)
    plt.tight_layout()
    plt.show()

    # Figure 4: Confusion Matrix (Separate)
    fig4 = plt.figure(figsize=(8, 6))
    cm = confusion_matrix(y_test, y_test_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True,
                annot_kws={"size": 16, "weight": "bold"},
                linewidths=2, linecolor='black')
    plt.title(f'{model_name} - Confusion Matrix', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Predicted Label', fontsize=13, fontweight='bold')
    plt.ylabel('True Label', fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.show()

    # Classification Report
    print(f"\n📋 Classification Report:")
    print("=" * 90)
    print(classification_report(y_test, y_test_pred, digits=4))
    print("=" * 90 + "\n")

    return {
        'model_name': model_name,
        'train_acc': train_accuracy,
        'test_acc': test_accuracy,
        'cv_acc': mean_cv_score,
        'train_loss': train_loss,
        'test_loss': test_loss,
        'train_roc_auc': train_roc_auc,
        'test_roc_auc': test_roc_auc,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'train_fpr': train_fpr,
        'train_tpr': train_tpr,
        'test_fpr': test_fpr,
        'test_tpr': test_tpr
    }


def create_final_comparison_table(accuracies_df):
    """
    Create comprehensive comparison table
    """
    print("\n" + "="*140)
    print(" " * 55 + "FINAL MODEL COMPARISON TABLE")
    print("="*140)

    print(accuracies_df.to_string(index=False, float_format=lambda x: f'{x:.4f}'))
    print("="*140)

    print("\n🏆 BEST MODELS BY METRIC:")
    print("-" * 80)
    print(f"  • Highest Test Accuracy:      {accuracies_df.loc[accuracies_df['Test Accuracy'].idxmax(), 'Model']:<30} ({accuracies_df['Test Accuracy'].max():.4f})")
    if accuracies_df['Test Loss'].sum() > 0:
        print(f"  • Lowest Test Loss:           {accuracies_df.loc[accuracies_df['Test Loss'][accuracies_df['Test Loss'] > 0].idxmin(), 'Model']:<30} ({accuracies_df['Test Loss'][accuracies_df['Test Loss'] > 0].min():.4f})")
    print(f"  • Highest Test ROC-AUC:       {accuracies_df.loc[accuracies_df['Test ROC-AUC'].idxmax(), 'Model']:<30} ({accuracies_df['Test ROC-AUC'].max():.4f})")
    print(f"  • Highest F1-Score:           {accuracies_df.loc[accuracies_df['F1-Score'].idxmax(), 'Model']:<30} ({accuracies_df['F1-Score'].max():.4f})")
    print(f"  • Highest CV Accuracy:        {accuracies_df.loc[accuracies_df['Mean CV Accuracy'].idxmax(), 'Model']:<30} ({accuracies_df['Mean CV Accuracy'].max():.4f})")

    print("="*140 + "\n")

    return accuracies_df


# ======================== MAIN EXECUTION ========================

# Add this after your existing preprocessing code in the original file:
# After you have X_train, X_test, y_train, y_test ready...


results = []

# Train all models
print("\n" + "🚀 "*30)
print("STARTING MODEL TRAINING AND EVALUATION")
print("🚀 "*30 + "\n")

# 1. Base Random Forest
print("\n🔹 [1/6] Training: Base RandomForest")
rf = RandomForestClassifier(random_state=42, n_estimators=100)
result = train_evaluate_model_complete(rf, X_train, y_train, X_test, y_test, "Base RandomForest")
results.append(result)

# 2. Tuned Random Forest
print("\n🔹 [2/6] Training: Tuned RandomForest")
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
print("Performing GridSearchCV...")
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid,
                          cv=3, n_jobs=-1, verbose=0, scoring='accuracy')
grid_search.fit(X_train, y_train)
best_rf = grid_search.best_estimator_
print(f"✓ Best parameters found: {grid_search.best_params_}")
result = train_evaluate_model_complete(best_rf, X_train, y_train, X_test, y_test, "Tuned RandomForest")
results.append(result)

# 3. Decision Tree
print("\n🔹 [3/6] Training: Decision Tree")
dt = DecisionTreeClassifier(random_state=42)
result = train_evaluate_model_complete(dt, X_train, y_train, X_test, y_test, "Decision Tree")
results.append(result)

# 4. SVC
print("\n🔹 [4/6] Training: Support Vector Classifier")
svc = SVC(probability=True, random_state=42)
result = train_evaluate_model_complete(svc, X_train, y_train, X_test, y_test, "Support Vector Classifier")
results.append(result)

# 5. Logistic Regression
print("\n🔹 [5/6] Training: Logistic Regression")
lr = LogisticRegression(max_iter=2000, random_state=42)
result = train_evaluate_model_complete(lr, X_train, y_train, X_test, y_test, "Logistic Regression")
results.append(result)

# 6. XGBoost
print("\n🔹 [6/6] Training: XGBoost")
xgb = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')
result = train_evaluate_model_complete(xgb, X_train, y_train, X_test, y_test, "XGBoost")
results.append(result)

# Create final comparison DataFrame
model_names = [r['model_name'] for r in results]

accuracies_df = pd.DataFrame({
    'Model': model_names,
    'Training Accuracy': Atrain,
    'Test Accuracy': Atest,
    'Mean CV Accuracy': Acv,
    'Training Loss': train_losses,
    'Test Loss': test_losses,
    'Train ROC-AUC': [r['train_roc_auc'] for r in results],
    'Test ROC-AUC': roc_auc_scores,
    'Precision': precision_scores,
    'Recall': recall_scores,
    'F1-Score': f1_scores
})

# Display final table
final_table = create_final_comparison_table(accuracies_df)

# Save results
accuracies_df.to_csv('complete_model_comparison_results.csv', index=False)
print("\n✅ Results saved to 'complete_model_comparison_results.csv'")
print("\n" + "🎉 "*30)
print("ALL MODELS TRAINING AND EVALUATION COMPLETED!")
print("🎉 "*30 + "\n")
```
